```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("Churn_Modelling.csv")
a
```

```python
a.isnull().sum()
```

```python
a.replace(["Male","Female"],[0,1],inplace=True)
```

```python
a
```

```python
a.Geography.value_counts()
```

```python
a.replace(["France","Germany","Spain"],[0,1,2],inplace=True)
```

```python
a
```

```python
x=a[["CreditScore","Geography","Gender","Age","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]
x
```

```python
y=a.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=150)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import BernoulliNB
```

```python
DTC=DecisionTreeClassifier()
KNN=KNeighborsClassifier()
RFC=RandomForestClassifier()
model=LogisticRegression()
BernoulliNB=BernoulliNB()
```

```python
DTC.fit(xtrain,ytrain)
```

```python
DTC.predict(xtest)
```

```python
DTC.score(xtest,ytest)
```

```python
KNN.fit(xtrain,ytrain)
```

```python
KNN.predict(xtest)
```

```python
KNN.score(xtest,ytest)
```

```python
RFC.fit(xtrain,ytrain)
```

```python
RFC.predict(xtest)
```

```python
RFC.score(xtest,ytest)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

```python
BernoulliNB.fit(xtrain,ytrain)
```

```python
BernoulliNB.predict(xtest)
```
